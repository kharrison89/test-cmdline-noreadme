### The code below will convert from Newtons (N) to kilograms (kg), and display the kg.

```python
n = 2519

kg = n/9.81

print(kg)
```

If you are too lazy to do the calculation yourself, the code aboove, or __[this website](https://www.cleancss.com/convert-units/units-of-weight/N/kg)__ will perform the conversion for you. 
<br>When using this conversion, there are some things you should keep in mind:
*  N represents force or weight, while kg are units of mass. 
*  This conversion only applies where g = 9.81 m/s^2 (i.e. Earth)
*  mass of a given object will be the same on any planet, while its weight will change according to the gravitational force
